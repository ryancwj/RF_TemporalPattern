```python
# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import rpy2.robjects as robjects
from scipy.stats import norm
```

```python
#Run IETD R Script
r = robjects.r('source("C:/Projects/Side Projects/Monte Carlo/Script/IETD.r")')
```

```python
#Import storm duration data
df = pd.read_csv(r"C:\Projects\Side Projects\Monte Carlo\Script\Output\Filtered_Rainfall_Characteristics.csv", usecols = ["Rainfall_Characteristics.Duration"])
x = df.squeeze().sort_values(ascending=False)

#Calculate mean and Standard deviation.
mean = np.mean(x)
sd = np.std(x)

#plot
plt.plot(x, norm.pdf(x,mean,sd))
plt.xlabel('Storm Duration')
plt.ylabel('Probability Density')
plt.title("Mean Storm Duration = " + str(round(mean,2)) + " hours", loc = "center")
plt.hlines(norm.pdf(mean,mean,sd),0,mean,linestyle ="dotted", color = "black")
plt.vlines(mean,0,norm.pdf(mean,mean,sd),linestyle ="dotted", color = "black")
plt.savefig(r"C:\Projects\Side Projects\Monte Carlo\Script\Output\Storm Duration Probability Distribution.png")
plt.show()
```
